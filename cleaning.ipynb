```python
import pandas as pd
```

```python
data_1 = pd.read_csv('dataset/emails.csv')
data_1.head()
```

```python
data_1['text'] = data_1['text'].str.replace('Subject:', '')
```

```python
data_1 = data_1.rename(columns={'spam': 'label'})
data_1
```

```python
data_1 = data_1.replace([1, 0], ['spam', 'not spam'])
data_1
```

```python
data_1.info()
```

```python
data_2 = pd.read_csv('dataset\spam_or_not_spam.csv')
data_2.head()
```

```python
data_2.info()
```

```python
data_2 = data_2.dropna(subset=['email'])
```

```python
data_2 = data_2.rename(columns={'email': 'text'})
```

```python
data_2 = data_2.replace([0, 1], ['not spam', 'spam'])
```

```python
data_2.isnull().sum()
```

```python
data_3 = pd.read_csv('dataset/spam.csv')
data_3.head()
```

```python
data_3 = data_3.rename(columns={'Message': 'text', 'Category': 'label'})
data_3 = data_3.replace(['ham'], ['not spam'])
data_3 = data_3[['text', 'label']]
data_3
```

```python
data_1
```

```python
data_2
```

```python
data_3
```

```python
data = pd.concat([data_1, data_2, data_3], ignore_index=True)
data
```

```python
data.label.unique()
```

```python
data.to_csv('dataset\cleaned_dataset.csv', index=False)
print('Your dataset has been cleaned, merged, and saved.')
```

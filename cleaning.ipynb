```python
import pandas as pd
```

```python
data_1 = pd.read_csv('dataset/email_spam.csv')
data_1.head()
```

```python
# Menghapus kolom yang tidak diperlukan
data_1 = data_1.drop('title', axis=1)
data_1.head()
```

```python
data_1.type.unique()
```

```python
data_1['text'] = data_1['text'].str.replace('\n', ' ')
data_1
```

```python
data_1.isnull().count()
```

```python
data_2 = pd.read_csv('dataset\emails.csv')
data_2.head()
```

```python
data_2 = data_2.rename(columns={'spam': 'type'})
data_2
```

```python
data_2 = data_2.replace([1, 0], ['spam', 'not spam'])
data_2
```

```python
data_2.isnull().count()
```

```python
data_3 = pd.read_csv('dataset/spam_or_not_spam.csv')
data_3.head()
```

```python
data_3.label.unique()
```

```python
data_3 = data_3.rename(columns={'label': 'type', 'email': 'text'})
data_3
```

```python
data_3 = data_3.replace([1, 0], ['spam', 'not spam'])
data_3
```

```python
data_3.isnull().count()
```

```python
data_4 = pd.read_csv('dataset/spam.csv', encoding='latin-1')
data_4.head()
```

```python
data_4 = data_4.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)
data_4.head()
```

```python
data_4 = data_4.rename(columns={'v1': 'type', 'v2': 'text'})
data_4
```

```python
data_4 = data_4[['text', 'type']]
data_4
```

```python
data_4 = data_4.replace(['spam', 'ham'], ['spam', 'not spam'])
data_4
```

```python
data_4.isnull().count()
```
